```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
# df2['text'] = df2['text'].apply(lambda x: x.split())
```

```python
stop_words = set(stopwords.words('english'))
```

```python
df2['text'] = df2['text'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in stop_words]))
```

```python
df2
```

```python
import re
import string
from bs4 import BeautifulSoup
```

```python
df2['text'] = df2['text'].replace(string.punctuation, '')  # Remove punctuation
```

```python
df2['text']
```

```python
df2['text'] = df2['text'].apply(lambda x: re.sub( r'[^\w\s]', '', x))  # Remove special characters
```

```python
df2
```

```python
df2['text'] = df2['text'].apply(lambda x: re.sub(r'http\S+', '', x))  # Remove URLs
```

```python
df2
```

```python
df2['sentiment'] = df2['sentiment'].replace({'positive': 2, 'neutral': 1, 'negative': 0})
```

```python
df2
```

```python
df2['text'] = df2['text'].astype(str)
```

```python
df2['combine'] = df2['text'] + ' ' + df2['sentiment'].astype(str)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
history = model.fit(train_x, train_y, epochs=30, validation_data=(test_x, test_y))
```

```python
results = model.evaluate(test_x, test_y, verbose=2)

for name, value in zip(model.metrics_names, results):
  print("%s: %.3f" % (name, value))
```
