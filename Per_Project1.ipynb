#### 1. Text Data Preprocessing

```python
import string
string.punctuation
```

```python
df['genre'] = df['genre'].astype(str)
df['type'] = df['type'].astype(str)
```

```python
df['genre'] = df['genre'].str.lower()
df['type'] = df['type'].str.lower()
```

```python
df[['type', 'genre']]
```

```python
exclude = string.punctuation
```

```python
# Remove tags and punctuations
def remove_punc(text):
  for char in exclude:
    text = text.replace(char, '')
  return text
```

```python
text2 = df['genre']
df['genre'] = remove_punc(text2)
```

```python
#Unique Values
```

```python
selected_columns = ['genre', 'type', 'rating']
df2 = df.drop_duplicates(selected_columns)
```

```python
df2
```

```python
df2['rating'] = df['rating'].astype(str)
```

```python
df2['combine'] = df2['genre']+' '+df2['type']+' '+df2['rating'] #Unique values combine features
```

```python
range(len(df2))
```

```python
#Creating an Index column
df2['index'] = range(len(df2))
```

```python
df2
```

#### 2. Converting text data into num values

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vector = TfidfVectorizer()
```

```python
f_vector = vector.fit_transform(df2['combine']) #Feature vectors
```

```python
print(f_vector)
```

```python
similarity = cosine_similarity(f_vector)
```

```python
print(similarity)
```

```python
similarity.shape
```

#### 3. Creating a list of anime names

```python
anime_name_list = df2['name'].to_list()
print(anime_name_list)
```

#### 4. Finding Close Match

```python
import difflib
```

```python
anime_name = 'hunter x hunter'
```

```python
find_close_match = difflib.get_close_matches(anime_name, anime_name_list)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
df[df.name == close_match].values[0]
```

```python
index_of_anime1 = df[df.name == close_match].values[0]
index_of_anime2 = index_of_anime1[0]
print(index_of_anime2)
```

#### 5. Getting Similar Anime

```python
similar_score = list(enumerate(similarity[index_of_anime2]))
print(similar_score)
```

#### 6. Anime Recommendations function

```python
def anime_recommendation():
  anime_name = input("Enter Anime Name: ")
  find_close_match = difflib.get_close_matches(anime_name, anime_name_list)
  close_match = find_close_match[0]
  index_of_anime1 = df[df.name == close_match].values[0] # All details are there because it doesn't have index column
  index_of_anime2 = index_of_anime1[0] # Only index of anime in dataset
  similar_score = list(enumerate(similarity[index_of_anime2]))
  similar_score_sort = sorted(similar_score, key = lambda x:x[1], reverse = True)
  for i in range(11):
    index = similar_score_sort[i][0]
    print(df2.iloc[index]['name'])
```

```python
anime_recommendation()
```
